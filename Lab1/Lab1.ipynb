# Q1

Generating a tally table

```python
# Generating random numbers
var = [random.randrange(1, 5) for _ in range(50)]
print(var)
```

```python
# Converting the list as pandas dataframe
data = pd.DataFrame({'class': var})
unique = data['class'].sort_values().unique()
```

```python
# Function to generate tally
def generate_tally(count):
    var = ''
    for i in range(count):
        if((i+1)%5 == 0):
            var = var + "\\ "
        else:
            var = "".join([var, '|'])
    return var
```

```python
print(generate_tally(13))
```

```python
# Final table containing the class, tally and frequency
mod_data = pd.DataFrame({'class': unique})
mod_data['tally'] = ''
mod_data['Frequency'] = 0
for i in mod_data['class']:
    mod_data['Frequency'][mod_data['class']==i] = data['class'].value_counts()[i]
    
mod_data['tally'] = mod_data['Frequency'].apply(generate_tally)
mod_data
```

# Q2

Need to find mean and median of marks obtained by given formula

```python
# Generting roll no
roll_no = []
for i in range(1,19):
    roll_no.append('CSE20D'+str(i) if(i>=10) else 'CSE20D0'+str(i))
```

```python
# Function to generate marks based on roll no
def get_marks(roll):
    num = int(roll[-2:])
    if(num%2==0):
        marks = 25+((num+8)%10)
    else:
        marks = 25+((num+7)%10)
    return marks
```

```python
# Generating a data set based on roll no and marks
data2 = pd.DataFrame({'Roll No': roll_no})
data2['Marks'] = 0
data2['Marks'] = data2['Roll No'].apply(get_marks)
data2
```

```python
Mean = data2['Marks'].mean()
Median = data2['Marks'].median()
print("Mean="+str(Mean)+" Median="+str(Median))
```

# Q8

Relative Grading

```python
# Importing data from excel
data8 = pd.read_excel("Marks.xlsx")
data8
```

```python
# Finding pass minimum
avg = data7['Total'].mean()
E_limit = avg/2
E_limit
```

```python
# Finding pass students mean
pass_count = 0
sum = 0

for i in data8['Total']:
    if(i>=E_limit):
        sum = sum + i
        pass_count = pass_count + 1

PassStudentsMean = sum/pass_count
PassStudentsMean
```

```python
X = PassStudentsMean - E_limit
X
```

```python
# Finding S cutoff
S_cutoff = data8['Total'].max() - 0.1 *(data8['Total'].max()-PassStudentsMean)
S_cutoff
```

```python
Y = S_cutoff - PassStudentsMean
Y
```

```python
print("PassStudentsMean = " + str(PassStudentsMean) + "\nX = "+str(X) + "\nY = "+str(Y))
```

```python
# Finding remaining grade cutoffs
A_cutoff = PassStudentsMean + Y * (5/8)
B_cutoff = PassStudentsMean + Y * (2/8)
C_cutoff = PassStudentsMean - X * (2/8)
D_cutoff = PassStudentsMean - X * (5/8)
print("S cutoff = "+str(S_cutoff)+"\nA cutoff = "+str(A_cutoff)+"\nB cutoff = "+str(B_cutoff)+"\nC cutoff = "+str(C_cutoff)+"\nD cutoff = "+str(D_cutoff)+"\nE cutoff = "+str(E_limit))
```

```python
# Function to assign grades based on cutoff and mark
def get_grade(mark):
    if(mark>=S_cutoff):
        return 'S'
    if(mark>=A_cutoff):
        return 'A'
    if(mark>=B_cutoff):
        return 'B'
    if(mark>=C_cutoff):
        return 'C'
    if(mark>=D_cutoff):
        return 'D'
    if(mark>=E_limit):
        return 'E'
    return 'F'
```

```python
# Applying grade for total given
data8['Grade'] = ''
data8['Grade'] = data8['Total'].apply(get_grade)
data8
```

```python
# Function to apply color to row based on grade
def row_style(row):
    color = ''
    if row.Grade == 'F':
        return pd.Series('background-color: red', row.index)
    if row.Grade == 'S':
        return pd.Series('background-color: lightgreen', row.index)
    if row.Grade == 'A':
        return pd.Series('background-color: yellow', row.index)
    if row.Grade == 'B':
        return pd.Series('background-color: blue', row.index)
    if row.Grade == 'C':
        return pd.Series('background-color: chocolate', row.index)
    if row.Grade == 'D':
        return pd.Series('background-color: pink', row.index)
    if row.Grade == 'E':
        return pd.Series('background-color: violet', row.index)

    
data8.style.apply(row_style, axis=1)
```

```python
# Displaying frequency of grades
grades_freq = data8.groupby(['Grade'])['Roll No'].count()
grades_freq
```
