Q2) a

```python
sorted_volumes = list(data2['Total Volume'].sort_values())
# print(sorted_volumes)
len_values = len(sorted_volumes)
print("Total Volume values: ", len_values)

no_of_bins = 250

bin_size = len_values//no_of_bins
print("Bin Size: ", bin_size)

bins = []

for i in range(250):
    bins.append(sorted_volumes[i: i+bin_size])
    
print(bins[0])

# Verification of length of each bin
s = 0
for i in range(len(bins)):
    s = s + len(bins[i])

# print("Avg of bin length: ", s//250)
```

Q2a) i) Bin Means

```python
sum1 = 0
bin_mean = []
for i in range(len(bins)):
    sum1 = sum(bins[i])
    avg = sum1//len(bins[i])
    
    bin_mean.append(avg)

# print(bin_mean)

smooth_bin_mean = []
for i in range(len(bin_mean)):
    tmp = []
    for j in range(bin_size):
        tmp.append(bin_mean[i])
        
    smooth_bin_mean.append(tmp)


print(smooth_bin_mean)
```

Q2a) ii) Bin Medians

```python
bin_median = []
for i in range(len(bins)):
    bin_median.append(median(bins[i]))

# print(bin_median)

smooth_bin_median = []
for i in range(len(bin_median)):
    tmp = []
    for j in range(bin_size):
        tmp.append(bin_median[i])
        
    smooth_bin_median.append(tmp)


print(smooth_bin_median)
```

Q2) a) iii: Bin Boundaries

```python
bin_boundaries = []
smooth_bin_boundaries = []

for i in range(len(bins)):
    bin_min = min(bins[i])
    bin_max = max(bins[i])
    tmp = []
    for j in range(len(bins[i])):
        diff_min = round(abs(bins[i][j]-bin_min), 2)
        diff_max = round(abs(bin_max-bins[i][j]), 2)
        
        if(diff_min < diff_max):
            tmp.append(bin_min)
        else:
            tmp.append(bin_max)
        # print(diff_min, diff_max)
    smooth_bin_boundaries.append(tmp)

print(smooth_bin_boundaries)
```
